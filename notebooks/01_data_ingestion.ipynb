```python
fund_master = datasets["fund_master"]

print(fund_master.columns.tolist()) 
```

```python
# Extract Codes 
master_codes = set(fund_master['amfi_code'])

nav_codes = set(datasets["nav_history"]['amfi_code']) 
```

```python
# Codes present in fund master but missing in NAV history 
missing_codes = master_codes - nav_codes

print("Missing Codes :", len(missing_codes))
print(sorted(list(missing_codes))) 
```

```python
# Codes present in NAV history but not in Fund Master
extra_codes = nav_codes - master_codes

print("Extra Codes :", len(extra_codes))

print(sorted(list(extra_codes)))   
```

```python
# Validation Result

if len(missing_codes) == 0:
    print("✅ All AMFI codes in fund_master exist in nav_history.")
else:
    print("❌ Some AMFI codes are missing from nav_history.") 
```

```python
# Percentage Coverage
coverage = (len(master_codes & nav_codes) / len(master_codes)) * 100

print(f"Coverage : {coverage:.2f}%") 
```
